````python
url3="https://api.github.com/users/jinverno/repos"
response3=requests.get(url3, auth = HTTPBasicAuth('jinverno', '0734204c0b15258bb11a167925989d856085b7e3'))
response3
````

````python
df3=pd.json_normalize(response3.json())
df3.head()
````

````python
repo_list=df3["commits_url"][df3.index]
repo_list.shape
````

````python
df4=pd.DataFrame([])
for index,i in enumerate(repo_list):
    url3=str(i[:-6])
    print(url3)
    response=requests.get(url3, auth = HTTPBasicAuth('jinverno', '0734204c0b15258bb11a167925989d856085b7e3'))
    df4=df4.append(pd.DataFrame(pd.json_normalize(response.json()))[["commit.author.name","commit.author.date","url"]],ignore_index=True)
print(df4.shape)
df4.head()
````

````python
df4_last2weeks=df4[(df4["commit.author.name"]=="jinverno") & (df4["commit.author.date"]>'2020-06-29')]
df4_last2weeks.reset_index(inplace=True)
print("Below are the",df4_last2weeks.shape[0] ,"commits done by jinverno overall repositories in the last 2 weeks")
df4_last2weeks
````
